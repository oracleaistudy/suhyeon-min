**08-1 합성곱 신경망의 구성 요소**

**합성곱 신경망 convolutional neural network CNN**   
- 이미지의 작은 부분 -> 전체로 보면서 의미를 찾아가는 모델   
- 왜 부분 -> 전체? : 정보량이 너무 커서 패턴 잡기 힘듦   
그래서 작은 특징 -> 큰 그림으로 점차 쌓아 올리는 구조 (convolutional)

**합성곱 convolution** : 필터를 이미지 위에 올려 연산 (곱셈 + 덧셈)   

*import keras*
*keras.layers.Conv2D(10, kernel_size=(3, 3), activation='relu', padding='same', strides=1)*      

**필터 filter (=커널 kernel)** : 이미지에 가중치를 더하는 장치 (3 * 3, 5 * 5를 많이 사용)      
- 처음엔 랜덤 -> 학습을 통해 스스로 만들어짐   
- 필터의 숫자로 용도가 달라짐

**특성 맵 feature map** : 합성곱 계산을 통해 얻은 output (input보다 크기가 작아짐)   

**패딩 padding** : input 배열 주위를 가상의 원소로 채우는 것   
-> 커널이 도장을 찍을 횟수를 늘려주어 output의 크기 조절  
예) (4,4) 크기의 input을 (6,6) 크기의 input으로 바꾸고 싶을 때 0으로 채우는 것   

**same padding** : input과 feature map 크기가 동일 (많이 사용)   
-> 모서리 픽셀은 합성곱 계산 수가 적을 수밖에 없는데 padding을 통해서 중앙부 픽셀과 합성곱 계산 수가 동일해짐    
**valid padding** : padding 없이 진행하는 것 -> feature map의 크기가 줄어들 수밖에 없음   

**stride** : 합성곱 연산에서 n칸 씩 이동하는 매개변수      

**pooling** : feature map의 가로세로 크기를 줄이는 역할, 개수는 아님 !   
예) (2,2,3) 크기의 특성 맵 -> (1,1,3) 크기로 줄임    

커널과 비슷하게 input위를 지나가면서 도장을 찍지만 pooling에는 가중치가 없음    

*keras.layers.MaxPooling2D(2, strides=2, padding='valid')*      

- max pooling : 도장을 찍은 영역에서 가장 큰 값을 고름 (MaxPooling2D)         
- average pooling : 도장을 찍은 영역의 값들의 평균값 (AveragePooling2D)     
**-> average pooling은 feature map에 있는 중요한 정보를 희석시킬 수 있어서 max pooling을 더 많이 사용**   

feature map이 여러 개라면 동일한 작업을 반복   
pooling은 겹치지 않고 이동 -> pooling 크기 만큼 이동   
예) pooling의 크기가 (2,2) 이면 2칸씩 이동   

*매개변수   
- 풀링 크기 : 대부분 2 (크기를 1/2 줄임), 가로세로 방향의 풀링 크기를 다르게 할 수 있음 (2,3)   
- strides : 기본값은 풀링의 크기   
- padding : 기본값은 'valid' -> 패딩x   
**pooling은 대체적으로 strides와 padding을 변경 하지 않는다**   

**stride를 크게하여 특성맵을 줄이는 것보다 풀링 층에서 크기를 줄이는 것이 더 좋은 성능을 냄 !!**    



**<합성곱 신경망의 전체 구조>**   
1. 합성곱층   
2. 풀링층   
3. 밀집층 (출력층)     

밀집층에 전달하려면 3차원 배열을 1차원으로 펼쳐야 함 (Flatten 클래스)      
예) (2,2,3) -> 12개의 원소를 가진 1차원 배열   

밀집층에 3개의 뉴런 -> 3개의 클래스를 분류하는 다중 분류 문제   

출력층에서 계산된 값은 소프트맥스 활성화 함수를 거쳐 최종 예측 확률이 됨
